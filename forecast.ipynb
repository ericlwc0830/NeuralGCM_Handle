```python
import datetime
import os
import sys
sys.path.append(os.path.abspath("./lib"))
from neuralgcm_handler import NeuralGCM, NeuralGCMResult
```

```python
# 指定Model
ngcm = NeuralGCM.from_name(1.4, "deterministic")

# 選擇初始時間、預報時間
initial_time = datetime.datetime(2023, 7, 25, 0, 0, 0)
output_per = datetime.timedelta(hours=1)
total_forecast_outputs = 1
ngcm.forecast_using_era5(initial_time, output_per, total_forecast_outputs)
```

```python
# 存檔
ngcm.save_forecast_to_netcdf4("output/output.nc")
```

```python
# 讀取結果
result = NeuralGCMResult.from_neuralgcm(ngcm)
```
